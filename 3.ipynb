```python
import pandas as pd
```

```python
df = pd.read_csv("forestfires.csv")
print(f"Loaded dataset with {len(df)} rows.\n")
```

```python
df.head()
```

```python
mapped = [(row["Month"], row["Temperature_Celsius"]) for _, row in df.iterrows()]
```

```python
result = {}
for month, temp in mapped:
    result.setdefault(month, []).append(temp)
```

```python
avg_temp = {m: sum(v)/len(v) for m, v in result.items()}
```

```python
print("üå°Ô∏è Average Temperature per Month:")
for m, t in avg_temp.items():
    print(f"{m}: {t:.2f}")
```

```python
top = df.groupby("Month")["Burned_Area_hectares"].mean().sort_values(ascending=False).head(5)
print("\nüî• Top 5 Months with Largest Burned Area:")
print(top)
```

```python
corr = df["Temperature_Celsius"].corr(df["Burned_Area_hectares"])
print(f"\nüìä Correlation between Temperature and Burned Area: {corr:.2f}")
```
